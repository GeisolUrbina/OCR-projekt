### EasyOCR pipeline

```python

import easyocr
from pdf2image import convert_from_path
from pathlib import Path
import numpy as np


pdf_path = "pdf_test.pdf"

# Konvertera PDF till bilder
print("Konverterar PDF till bilder...")
pages = convert_from_path(pdf_path, dpi=250)

print(f"Antal sidor: {len(pages)}")

# Initiera EasyOCR
print("Laddar EasyOCR-modeller...")
reader = easyocr.Reader(['sv', 'en'], gpu=True)

#  Kör OCR på alla sidor
print("Kör OCR på PDF-sidor...")
all_text = ""

for i, page in enumerate(pages):
    print(f"OCR sida {i+1}/{len(pages)}...")
    page_np = np.array(page)
    result = reader.readtext(page_np, detail=1, paragraph=True)
    for item in result:
        all_text += item[1] + "\n"

# Exportera till markdown
output_file = "easyocr_output.md"
Path(output_file).write_text(all_text, encoding="utf-8")

print(f"\n Text sparad i: {output_file}\n")
print("Förhandsvisning av resultat:")
print(all_text[:2000])
```

```python
# OCR BLOCKS för quality report

ocr_blocks = []

for i, page in enumerate(pages):
    page_np = np.array(page)
    result = reader.readtext(page_np, detail=1, paragraph=False)
    for bbox, text, conf in result:
        ocr_blocks.append({
            "page": i+1,
            "bbox": [[float(x), float(y)] for (x,y) in bbox],
            "text": text,
            "confidence": float(conf)
        })

print(f"OCR-block extraherade: {len(ocr_blocks)}")
```

```python
import re
import numpy as np
from difflib import SequenceMatcher
from jiwer import wer


# OCR SCORE
def ocr_score(ocr_blocks):
    if not ocr_blocks:
        return 0.0

    confs = [b["confidence"] for b in ocr_blocks]
    return float(np.mean(confs))


# LAYOUT SCORE

def layout_score(ocr_blocks):
    if not ocr_blocks:
        return 0.0

    def area(bbox):
        xs = [p[0] for p in bbox]
        ys = [p[1] for p in bbox]
        return (max(xs) - min(xs)) * (max(ys) - min(ys))

    areas = [area(b["bbox"]) for b in ocr_blocks]
    mean = np.mean(areas)
    std = np.std(areas)

    if mean == 0:
        return 0.0

    # låg variation = stabil layout = högt score
    return float(max(0, 1 - std / mean))


# PARSE SCORE (Docling markdown)

def parse_score(md_text):
    lines = md_text.split("\n")
    clean = 0
    total = len(lines)

    for l in lines:
        l = l.strip()
        # räkna rader som verkar "ren text"
        if len(l) > 2 and not re.match(r"^[\W_]+$", l):
            clean += 1

    if total == 0:
        return 0.0

    return clean / total


```

```python

# MATCH SCORE (Docling vs OCR)

def match_score(docling_md, ocr_blocks):
    # kombinerad likhet mellan Docling + OCR
    ocr_text = " ".join([b["text"] for b in ocr_blocks])

    # rensa text
    def clean(t):
        t = re.sub(r"[^A-Za-zÅÄÖåäö0-9\s]", " ", t)
        t = re.sub(r"\s+", " ", t).strip()
        return t.lower()

    doc = clean(docling_md)
    ocr = clean(ocr_text)

    if len(doc) == 0 or len(ocr) == 0:
        return 0.0

    # char similarity
    char_sim = SequenceMatcher(None, doc, ocr).ratio()

    # token similarity (Jaccard)
    dt = set(doc.split())
    ot = set(ocr.split())
    jaccard = len(dt & ot) / max(1, len(dt | ot))

    # word overlap
    overlap = len(dt & ot) / max(1, len(dt))

    # WER (ju lägre desto bättre → vi vänder den)
    try:
        wer_score = (1 - wer(doc, ocr))
    except:
        wer_score = 0.0

    # slutligt match score (vägda delar)
    final = 0.4 * char_sim + 0.3 * jaccard + 0.2 * overlap + 0.1 * wer_score
    return float(final)
```

```python

# QUALITY REPORT BUILDER

def build_quality_report(docling_md, ocr_blocks):
    return {
        "ocr_score": ocr_score(ocr_blocks),
        "layout_score": layout_score(ocr_blocks),
        "parse_score": parse_score(docling_md),
        "match_score": match_score(docling_md, ocr_blocks),
        "total_score": float(
            0.25 * ocr_score(ocr_blocks) +
            0.25 * layout_score(ocr_blocks) +
            0.25 * parse_score(docling_md) +
            0.25 * match_score(docling_md, ocr_blocks)
        )
    }
```
